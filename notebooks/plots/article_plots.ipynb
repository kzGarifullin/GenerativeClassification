```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import umap
import plotly.express as px
import plotly.io as pio

features_umap = np.load('/content/gan-features-cifar-10.npy')
y = np.load('/content/gan-labels-cifar-10.npy')

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(features_umap)

umap_3d = umap.UMAP(n_components=3, random_state=1)
X_umap_3d = umap_3d.fit_transform(X_scaled)

df = pd.DataFrame(X_umap_3d, columns=['UMAP Component 1', 'UMAP Component 2', 'UMAP Component 3'])
df['label'] = y

df_sorted = df.sort_values(by='label')

colors = ['#0d0887', '#46039f', '#7201a8', '#9c179e', '#bd3786', '#d8576b', '#ed7953', '#fb9f3a', '#fdca26', '#f0f921']

fig = px.scatter_3d(df_sorted, x='UMAP Component 1', y='UMAP Component 2', z='UMAP Component 3',
                    color=df_sorted['label'].astype(str),
                    color_discrete_sequence=colors)

fig.update_layout(scene=dict(
                    xaxis_title='UMAP Component 1',
                    yaxis_title='UMAP Component 2',
                    zaxis_title='UMAP Component 3'),
                  margin=dict(l=0, r=0, b=0, t=40))

fig.show()
```
